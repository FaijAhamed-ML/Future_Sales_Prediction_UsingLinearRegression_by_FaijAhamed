# Future Sales Prediction with Machine Learning

```python
import pandas as pd
```

```python
data = pd.read_csv("advertising.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

## Future Sales Prediction Model

```python
import numpy as np
x = np.array(data.drop(columns=["Sales"]))
y = np.array(data["Sales"])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y,
                                                test_size=0.2, 
                                                random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

```python
#features = [[TV, Radio, Newspaper]]
features = np.array([[230.1, 37.8, 69.2]])
y_pred = model.predict(features)
print("Predicted Sales:", y_pred)
```

```python
#retrive the coefficients and intercept of the model
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
#evaluate the model using r squared error and r2 score
print("R-squared (training):", model.score(xtrain, ytrain))
print("R-squared (testing):", model.score(xtest, ytest))
```

```python
#generate output from test data
y_pred_test = model.predict(xtest)
print("Predicted Sales for Test Data:", y_pred_test)
score = model.score(xtest, ytest)
print("R-squared Score (testing):", score)
```

```python
#evaluate the model using mean squared error
from sklearn.metrics import mean_squared_error
print("Mean Squared Error (testing):", mean_squared_error(ytest, y_pred_test))
```

```python
#evaluate root mean squared error
from sklearn.metrics import mean_squared_error
print("Root Mean Squared Error (testing):", np.sqrt(mean_squared_error(ytest, model.predict(xtest))))
```
